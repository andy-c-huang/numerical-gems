```python
import numpy, scipy
import matplotlib.pyplot as plt
```

```python
f = lambda x : numpy.exp(x)/(numpy.power(numpy.cos(x),3) + numpy.power(numpy.sin(x),3))
```

```python
xvalues = numpy.linspace(-numpy.pi/4,numpy.pi/2,10000)
plt.plot(xvalues, [f(x) for x in xvalues])
plt.ylim([0,6])
```

```python
hvalues = [1E-1, 1E-2, 1E-3, 1E-4, 1E-5, 1E-6]
xvalues = [numpy.linspace(-numpy.pi/4,numpy.pi/2,int(numpy.ceil(1./h))) for h in hvalues]
midpoints = [(xvalues[i][1:]+xvalues[i][:-1])/2 for i in range(len(hvalues))]
centered_diff = [(f(xvalues[i][1:]) - f(xvalues[i][:-1]))/hvalues[i] for i in range(len(hvalues))]
for i in range(len(hvalues)):
    plt.plot(midpoints[i], centered_diff[i], label=str(hvalues[i]))
plt.xlim([.2,.3])
plt.ylim([4,7])
plt.legend()
```

```python
hvalues_complex = [1E-1, 1E-2, 1E-3, 1E-4, 1E-5, 1E-6]
xvalues = [numpy.linspace(-numpy.pi/4,numpy.pi/2,int(numpy.ceil(1./h))) for h in hvalues_complex]
complex_step_approx = [numpy.imag(f(xvalues[i]+hvalues_complex[i]*1j))/hvalues_complex[i] \
                       for i in range(len(hvalues_complex))]
for i in range(len(hvalues_complex)):
    plt.plot(xvalues[i], complex_step_approx[i])
plt.xlim([.2,.3])
plt.ylim([0,4])
```
